# Multivariate Distributions

### Joint Distributions 

```python
# Imports 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Joint Distribution

# Define the possible outcomes for each die (1 to 6)
dice_outcomes = np.arange(1, 7)

# Create a grid for the joint distribution (each combination of the two dice)
joint_prob = np.ones((6, 6)) / 36  # All combinations have equal probability, 1/36

# Create a heatmap of the joint probability distribution
plt.figure(figsize=(8, 6))
sns.heatmap(joint_prob, annot=True, cmap="Blues", xticklabels=dice_outcomes, yticklabels=dice_outcomes)
plt.title("Joint Probability Distribution for Rolling Two Dice")
plt.xlabel("Outcome of Die 1")
plt.ylabel("Outcome of Die 2")
plt.show()

# Example: Calculate the probability that the sum of the two dice equals 7
sum_is_7 = np.array([[1, 6], [2, 5], [3, 4], [4, 3], [5, 2], [6, 1]])
prob_sum_7 = len(sum_is_7) / 36  # There are 6 combinations that result in a sum of 7
print(f"Probability that the sum of the two dice is 7: {prob_sum_7:.4f}")
```

### Conditional Distribution

```python
# Define possible outcomes for the second die
dice_outcomes = np.arange(1, 7)

# Since the first die is fixed at 5, the second die has equal probabilities
conditional_prob = np.ones(6) / 6  # Each outcome (1-6) has a 1/6 chance

# Plot the conditional distribution of the second die given that the first die is 5
plt.figure(figsize=(6, 4))
sns.barplot(x=dice_outcomes, y=conditional_prob, palette="Blues_d")
plt.title("Conditional Distribution: Second Die (Given First Die = 5)")
plt.xlabel("Outcome of Second Die")
plt.ylabel("Probability")
plt.show()

# Example: Probability of rolling a 4 on the second die given the first die is 5
prob_second_die_is_4 = conditional_prob[3]  # Index 3 corresponds to outcome 4
print(f"Probability that the second die is 4, given the first die is 5: {prob_second_die_is_4:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set the Seaborn theme to a dark style
sns.set_theme(style="darkgrid", palette="dark")

# Step 1: Generate 960 random numbers ranging from 1 to 1000
random_numbers = np.random.randint(1, 1001, size=960)

# Calculate properties for the random numbers
mean_random_numbers = np.mean(random_numbers)
variance_random_numbers = np.var(random_numbers)

# Print properties of the original 960 random numbers
print(f"Mean of 960 Random Numbers: {mean_random_numbers}")
print(f"Variance of 960 Random Numbers: {variance_random_numbers}")

# Step 2: Select 30 random samples, each consisting of 25 numbers and calculate their means
sample_means = []
for _ in range(30):
    sample = np.random.choice(random_numbers, size=25, replace=False)
    sample_means.append(np.mean(sample))

# Calculate properties of the sample means
mean_sample_means = np.mean(sample_means)
variance_sample_means = np.var(sample_means)

# Print properties of the sample means
print(f"\nMean of Sample Means (30 samples of 25 numbers): {mean_sample_means}")
print(f"Variance of Sample Means (30 samples of 25 numbers): {variance_sample_means}")

# Additional note: Number of samples, sample size
print(f"\nNumber of samples taken: 30")
print(f"Size of each sample: 25")

# Step 3: Plot the frequency distribution of the 960 random numbers with a cool blue theme
plt.figure(figsize=(8, 6))
sns.histplot(random_numbers, bins=30, kde=True, color='darkblue')
plt.title("Frequency Distribution of 960 Random Numbers", color='lightblue')
plt.xlabel("Number", color='lightblue')
plt.ylabel("Frequency", color='lightblue')
plt.show()

# Step 4: Plot the distribution of the sample means with a cool blue theme
plt.figure(figsize=(8, 6))
sns.histplot(sample_means, bins=10, kde=True, color='darkblue')
plt.title("Distribution of Sample Means (30 samples of 25 numbers)", color='lightblue')
plt.xlabel("Sample Mean", color='lightblue')
plt.ylabel("Frequency", color='lightblue')
plt.show()
```

#### Increasing the sample size and number of samples ...

```python
# Step 1: Increase the sample size to 100 and the number of samples to 100
extended_sample_means = []
for _ in range(100):
    extended_sample = np.random.choice(random_numbers, size=100, replace=False)
    extended_sample_means.append(np.mean(extended_sample))

# Step 2: Calculate properties of the extended sample means
mean_extended_sample_means = np.mean(extended_sample_means)
variance_extended_sample_means = np.var(extended_sample_means)

# Print properties of the extended sample means
print(f"\nMean of Extended Sample Means (100 samples of 100 numbers): {mean_extended_sample_means}")
print(f"Variance of Extended Sample Means (100 samples of 100 numbers): {variance_extended_sample_means}")

# Step 3: Plot the distribution of the extended sample means with the same cool blue theme
plt.figure(figsize=(8, 6))
sns.histplot(extended_sample_means, bins=10, kde=True, color='darkblue')
plt.title("Distribution of Extended Sample Means (100 samples of 100 numbers)", color='lightblue')
plt.xlabel("Sample Mean", color='lightblue')
plt.ylabel("Frequency", color='lightblue')
plt.show()
```

```python
from scipy.stats import multivariate_normal
# Set means and covariance
mu = [0, 0]  # mean of X and Y
sigma = [[1, 0.8], [0.8, 1]]  # covariance matrix

# Create grid
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
pos = np.dstack((X, Y))

# Create multivariate normal
rv = multivariate_normal(mean=mu, cov=sigma)
Z = rv.pdf(pos)

# Plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111)
contour = ax.contourf(X, Y, Z, cmap='viridis')
ax.set_title('Multivariate Normal Distribution (correlated)')
ax.set_xlabel('X')
ax.set_ylabel('Y')
plt.colorbar(contour)
plt.grid(True)
plt.show()
```
