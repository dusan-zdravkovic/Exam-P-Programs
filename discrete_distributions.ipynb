# Discrete Probability Distributions

```python
# Imports
import numpy as np
import math as m
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import randint
```

### Discrete Uniform 

```python

# Ex. dice roll with 6 values
a = 1 # lower bound parameter 
b = 6 # upper bound parameter 

# Generate the possible outcomes
x_values = np.arange(a, b+1) # hover over for function help

# PMF - Probabilty Mass Function
pmf = randint.pmf(x_values, a, b+1) # b+1 because we need 6 to be counted

# P(X=x) = a/(b-a+1) for uniform discrete 
# Plot the PMF using seaborn
plt.figure(figsize=(4, 3))
sns.barplot(x=x_values, y=pmf, palette="Blues_d")
plt.title(f"Discrete Uniform Distribution PMF (a={a}, b={b})")
plt.xlabel("X values (outcomes)")
plt.ylabel("PMF (Probability)")
plt.show()


```

```python
# Calculating the mean and variance 
mean = (a+b)/2
n = b - a + 1 # number of values in domain 
variance = (n**2 - 1)/12

print("The mean and variance is ", mean,"and ", variance )
```
