# Discrete Probability Distributions

```python
# Imports
import numpy as np
import math as m
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import randint
```

### Discrete Uniform 

```python

# Ex. dice roll with 6 values
a = 1 # lower bound parameter 
b = 6 # upper bound parameter 

# Generate the possible outcomes
x_values = np.arange(a, b+1) # hover over for function help

# PMF - Probabilty Mass Function
pmf = randint.pmf(x_values, a, b+1) # b+1 because we need 6 to be counted

# P(X=x) = a/(b-a+1) for uniform discrete 
# Plot the PMF using seaborn
plt.figure(figsize=(4, 3))
sns.barplot(x=x_values, y=pmf, palette="Blues_d")
plt.title(f"Discrete Uniform Distribution PMF (a={a}, b={b})")
plt.xlabel("X values (outcomes)")
plt.ylabel("PMF (Probability)")
plt.show()


```

```python
# Calculating the mean and variance 
mean = (a+b)/2
n = b - a + 1 # number of values in domain 
variance = (n**2 - 1)/12

print("The mean and variance is ", mean,"and ", variance )
```

### Bernouli & Binomial 

```python
from scipy.stats import bernoulli
# Parameters for the Bernoulli distribution
p = 0.5  # Probability of heads (success)

# Generate the possible outcomes: 0 for tails (failure), 1 for heads (success)
x_values = [0, 1]

# Calculate the PMF using scipy's bernoulli
pmf = bernoulli.pmf(x_values, p)

# Plot the PMF using seaborn
plt.figure(figsize=(4, 3))
sns.barplot(x=x_values, y=pmf, palette="Greens_d")
plt.title(f"Bernoulli Distribution PMF (p={p})")
plt.xlabel("X values (outcomes)")
plt.ylabel("PMF (Probability)")
plt.xticks([0, 1], ['Tails (0)', 'Heads (1)'])
plt.show()

# Calculating the mean and variance for Bernoulli
mean = p
variance = p * (1 - p)

print(f"The mean and variance of Bernoulli distribution are {mean} and {variance} respectively.")
```

```python
from scipy.stats import binom
# Parameters for the Binomial distribution
n = 10  # Number of trials (e.g., 10 coin flips)
p = 0.5  # Probability of heads (success)

# Generate the possible outcomes (number of heads) from 0 to n
x_values = np.arange(0, n + 1)

# Calculate the PMF using scipy's binom
pmf = binom.pmf(x_values, n, p)

# Plot the PMF using seaborn
plt.figure(figsize=(8, 5))
sns.barplot(x=x_values, y=pmf, palette="Purples_d")
plt.title(f"Binomial Distribution PMF (n={n}, p={p})")
plt.xlabel("Number of heads (successes)")
plt.ylabel("PMF (Probability)")
plt.show()

# Calculating the mean and variance for Binomial
mean = n * p
variance = n * p * (1 - p)

print(f"The mean and variance of the Binomial distribution are {mean} and {variance} respectively.")
```

### Hypergeometric 

```python
from scipy.stats import hypergeom

# Parameters for the hypergeometric distribution
N = 10  # Total number of balls (4 blue + 6 green)
m = 4   # Number of blue balls (successes in population)
n = 3   # Number of balls drawn (sample size)
k_values = np.arange(0, n + 1)  # Possible number of blue balls drawn

# Calculate the PMF (probability mass function)
pmf = hypergeom.pmf(k_values, N, m, n)

# Plot the PMF using seaborn
plt.figure(figsize=(8, 5))
sns.barplot(x=k_values, y=pmf, palette="Blues_d")
plt.title(f"Hypergeometric Distribution PMF\n(N={N}, m={m}, n={n})")
plt.xlabel("Number of blue balls drawn (k)")
plt.ylabel("PMF (Probability)")
plt.show()

# Specific probability: P(X = 2) where we draw exactly 2 blue balls
prob_2_blue = hypergeom.pmf(2, N, m, n)
print(f"Probability of drawing exactly 2 blue balls: {prob_2_blue}")
```

```python
from scipy.stats import geom

# Parameters for the Geometric distribution
p = 0.5  # Probability of getting heads (success)
k_values = np.arange(1, 11)  # We will compute for k = 1 to 10 trials

# Calculate the PMF using scipy's geom
pmf = geom.pmf(k_values, p)

# Plot the PMF using seaborn
plt.figure(figsize=(8, 5))
sns.barplot(x=k_values, y=pmf, palette="Oranges_d")
plt.title(f"Geometric Distribution PMF (p={p})")
plt.xlabel("Trial number (k)")
plt.ylabel("PMF (Probability)")
plt.show()

# Calculate the mean and variance for the Geometric distribution
mean = 1 / p
variance = (1 - p) / (p ** 2)

print(f"The mean is {mean} and the variance is {variance}.")
```
