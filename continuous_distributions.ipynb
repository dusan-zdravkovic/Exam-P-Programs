```python
# Imports
import numpy as np
import math as m
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from scipy.stats import uniform

# Define the parameters of the uniform distribution (waiting time in minutes)
a = 0        # Lower bound (0 minutes)
b = 30       # Upper bound (30 minutes)

# Generate a range of x values for the PDF and CDF
x_values = np.linspace(a, b, 1000)

# Calculate the PDF and CDF using scipy's uniform distribution
pdf = uniform.pdf(x_values, a, b - a)
cdf = uniform.cdf(x_values, a, b - a)

# Plot the PDF using seaborn
plt.figure(figsize=(8, 5))
sns.lineplot(x=x_values, y=pdf, color="blue", label="PDF")
plt.fill_between(x_values, pdf, color="lightblue", alpha=0.5)
plt.title(f"PDF of Continuous Uniform Distribution\nInterval = [{a}, {b}] (Bus Waiting Time)")
plt.xlabel("Waiting Time (minutes)")
plt.ylabel("PDF (Probability Density)")
plt.legend()
plt.show()

# Plot the CDF using seaborn
plt.figure(figsize=(8, 5))
sns.lineplot(x=x_values, y=cdf, color="green", label="CDF")
plt.title(f"CDF of Continuous Uniform Distribution\nInterval = [{a}, {b}] (Bus Waiting Time)")
plt.xlabel("Waiting Time (minutes)")
plt.ylabel("CDF (Cumulative Probability)")
plt.legend()
plt.show()

# Calculating the expected value (mean) and variance for the uniform distribution
mean = (a + b) / 2
variance = ((b - a) ** 2) / 12

print(f"Expected waiting time (mean) = {mean} minutes")
print(f"Variance of waiting time = {variance:.2f} minutes^2")

# Example: Probability of waiting less than 10 minutes
prob_less_than_10 = uniform.cdf(10, a, b - a)
print(f"Probability of waiting less than 10 minutes = {prob_less_than_10:.2f}")
```
