```python
# Imports
import numpy as np
import math as m
import matplotlib.pyplot as plt
import seaborn as sns
```

### Continuous Uniform

```python
from scipy.stats import uniform

# Define the parameters of the uniform distribution (waiting time in minutes)
a = 0        # Lower bound (0 minutes)
b = 30       # Upper bound (30 minutes)

# Generate a range of x values for the PDF and CDF
x_values = np.linspace(a, b, 1000)

# Calculate the PDF and CDF using scipy's uniform distribution
pdf = uniform.pdf(x_values, a, b - a)
cdf = uniform.cdf(x_values, a, b - a)

# Plot the PDF using seaborn
plt.figure(figsize=(8, 5))
sns.lineplot(x=x_values, y=pdf, color="blue", label="PDF")
plt.fill_between(x_values, pdf, color="lightblue", alpha=0.5)
plt.title(f"PDF of Continuous Uniform Distribution\nInterval = [{a}, {b}] (Bus Waiting Time)")
plt.xlabel("Waiting Time (minutes)")
plt.ylabel("PDF (Probability Density)")
plt.legend()
plt.show()

# Plot the CDF using seaborn
plt.figure(figsize=(8, 5))
sns.lineplot(x=x_values, y=cdf, color="green", label="CDF")
plt.title(f"CDF of Continuous Uniform Distribution\nInterval = [{a}, {b}] (Bus Waiting Time)")
plt.xlabel("Waiting Time (minutes)")
plt.ylabel("CDF (Cumulative Probability)")
plt.legend()
plt.show()

# Calculating the expected value (mean) and variance for the uniform distribution
mean = (a + b) / 2
variance = ((b - a) ** 2) / 12

print(f"Expected waiting time (mean) = {mean} minutes")
print(f"Variance of waiting time = {variance:.2f} minutes^2")

# Example: Probability of waiting less than 10 minutes
prob_less_than_10 = uniform.cdf(10, a, b - a)
print(f"Probability of waiting less than 10 minutes = {prob_less_than_10:.2f}")
```

### Exponential 

```python
from scipy.stats import expon

# Define the rate parameter (lambda), we use the one we calculated earlier
lambda_ = 0.017325  # Rate parameter for the exponential distribution

# Generate a range of x values (time between bus arrivals)
x_values = np.linspace(0, 100, 1000)

# Calculate the PDF and CDF using scipy's expon distribution
pdf = expon.pdf(x_values, scale=1/lambda_)  # scale = 1/lambda
cdf = expon.cdf(x_values, scale=1/lambda_)

# Plot the PDF using seaborn
plt.figure(figsize=(8, 5))
sns.lineplot(x=x_values, y=pdf, color="blue", label="PDF")
plt.fill_between(x_values, pdf, color="lightblue", alpha=0.5)
plt.title(f"PDF of Exponential Distribution\n(lambda = {lambda_})")
plt.xlabel("Time between bus arrivals (minutes)")
plt.ylabel("PDF (Probability Density)")
plt.legend()
plt.show()

# Plot the CDF using seaborn
plt.figure(figsize=(8, 5))
sns.lineplot(x=x_values, y=cdf, color="green", label="CDF")
plt.title(f"CDF of Exponential Distribution\n(lambda = {lambda_})")
plt.xlabel("Time between bus arrivals (minutes)")
plt.ylabel("CDF (Cumulative Probability)")
plt.legend()
plt.show()

# Calculate the probability of waiting more than 50 minutes using the CDF
prob_greater_than_50 = 1 - expon.cdf(50, scale=1/lambda_)
print(f"Probability of waiting more than 50 minutes: {prob_greater_than_50:.4f}")
```

```python
from scipy.stats import gamma

# Define the parameters of the Gamma distribution
alpha = 3    # Shape parameter (alpha)
theta = 4    # Scale parameter (theta)
mean = alpha * theta  # Mean of the Gamma distribution
variance = alpha * theta**2  # Variance of the Gamma distribution

# Generate a range of x values (lifetime of the machine in years)
x_values = np.linspace(0, 50, 1000)

# Calculate the PDF and CDF using scipy's gamma distribution
pdf = gamma.pdf(x_values, alpha, scale=theta)
cdf = gamma.cdf(x_values, alpha, scale=theta)

# Plot the PDF 
plt.figure(figsize=(8, 5))
sns.lineplot(x=x_values, y=pdf, color="blue", label="PDF")
plt.fill_between(x_values, pdf, color="lightblue", alpha=0.5)
plt.title(f"PDF of Gamma Distribution\n(alpha = {alpha}, theta = {theta})")
plt.xlabel("Lifetime of machine (years)")
plt.ylabel("PDF (Probability Density)")
plt.legend()
plt.show()

# Plot the CDF 
plt.figure(figsize=(8, 5))
sns.lineplot(x=x_values, y=cdf, color="green", label="CDF")
plt.title(f"CDF of Gamma Distribution\n(alpha = {alpha}, theta = {theta})")
plt.xlabel("Lifetime of machine (years)")
plt.ylabel("CDF (Cumulative Probability)")
plt.legend()
plt.show()

# Print the calculated mean and variance
print(f"Mean of the Gamma distribution: {mean:.2f} years")
print(f"Variance of the Gamma distribution: {variance:.2f} years^2")

# Example: Probability that the machine survives more than 20 years
prob_greater_than_20 = 1 - gamma.cdf(20, alpha, scale=theta)
print(f"Probability that the machine survives more than 20 years: {prob_greater_than_20:.4f}")
```

### Normal Distribution 

```python
from scipy.stats import norm

# Define the parameters for the normal distribution
mu = 175  # Mean height in cm
sigma = 10  # Standard deviation in cm

# Generate a range of x values (heights in cm)
x_values = np.linspace(mu - 4 * sigma, mu + 4 * sigma, 1000)

# Calculate the PDF (Probability Density Function)
pdf = norm.pdf(x_values, mu, sigma)

# Calculate the CDF (Cumulative Distribution Function)
cdf = norm.cdf(x_values, mu, sigma)

# Plot the PDF using seaborn
plt.figure(figsize=(8, 5))
sns.lineplot(x=x_values, y=pdf, color="blue", label="PDF")
plt.title(f"PDF of Male Heights (mu = {mu} cm, sigma = {sigma} cm)")
plt.xlabel("Height (cm)")
plt.ylabel("PDF (Probability Density)")
plt.legend()
plt.show()

# Plot the CDF using seaborn
plt.figure(figsize=(8, 5))
sns.lineplot(x=x_values, y=cdf, color="green", label="CDF")
plt.title(f"CDF of Male Heights (mu = {mu} cm, sigma = {sigma} cm)")
plt.xlabel("Height (cm)")
plt.ylabel("CDF (Cumulative Probability)")
plt.legend()
plt.show()

# Example: Probability that a randomly selected male is taller than 185 cm
prob_taller_than_185 = 1 - norm.cdf(185, mu, sigma)
print(f"Probability that a male is taller than 185 cm: {prob_taller_than_185:.4f}")
```

```python
from scipy.stats import lognorm

# Define the parameters for the lognormal distribution
mu = 3.786  # Adjusted mean of the logarithm of stock price
sigma = 0.5  # Standard deviation of the logarithm of stock price

# Generate a range of x values (stock prices)
x_values = np.linspace(1, 200, 1000)

# Calculate the PDF (Probability Density Function)
pdf = lognorm.pdf(x_values, sigma, scale=np.exp(mu))

# Calculate the CDF (Cumulative Distribution Function)
cdf = lognorm.cdf(x_values, sigma, scale=np.exp(mu))

# Plot the PDF using seaborn
plt.figure(figsize=(8, 5))
sns.lineplot(x=x_values, y=pdf, color="blue", label="PDF")
plt.fill_between(x_values, pdf, color="lightblue", alpha=0.5)
plt.title(f"PDF of Stock Prices (lognormal with mu = {mu:.3f}, sigma = {sigma})")
plt.xlabel("Stock Price ($)")
plt.ylabel("PDF (Probability Density)")
plt.legend()
plt.show()

# Plot the CDF using seaborn
plt.figure(figsize=(8, 5))
sns.lineplot(x=x_values, y=cdf, color="green", label="CDF")
plt.title(f"CDF of Stock Prices (lognormal with mu = {mu:.3f}, sigma = {sigma})")
plt.xlabel("Stock Price ($)")
plt.ylabel("CDF (Cumulative Probability)")
plt.legend()
plt.show()

# Example: Probability that the stock price is greater than $50
prob_greater_than_50 = 1 - lognorm.cdf(50, sigma, scale=np.exp(mu))
print(f"Probability that the stock price is greater than $50: {prob_greater_than_50:.4f}")
```

### Beta

```python
from scipy.stats import beta

# Parameters for the Beta distribution (based on the rain example)
alpha = 4  # 4 days of rain
beta_param = 3  # 3 days of no rain

# Generate a range of x values between 0 and 1 (possible probabilities of rain)
x_values = np.linspace(0, 1, 1000)

# Calculate the PDF of the Beta distribution
pdf = beta.pdf(x_values, alpha, beta_param)

# Plot the PDF using seaborn
plt.figure(figsize=(8, 5))
sns.lineplot(x=x_values, y=pdf, color="blue", label=f"Beta PDF (alpha={alpha}, beta={beta_param})")
plt.fill_between(x_values, pdf, color="lightblue", alpha=0.5)
plt.title(f"Beta Distribution PDF - Chance of Rain (alpha={alpha}, beta={beta_param})")
plt.xlabel("Probability of Rain")
plt.ylabel("PDF (Probability Density)")
plt.legend()
plt.show()

# Example: Calculate the mean of the Beta distribution (the estimated probability of rain)
mean = alpha / (alpha + beta_param)
print(f"Estimated probability of rain tomorrow: {mean:.4f}")
```
