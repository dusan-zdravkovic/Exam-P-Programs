```python
import random
import matplotlib.pyplot as plt

# Parameters
num_trials = 100
bet_number = 36
payout_multiplier = 35  # Roulette payout for a single number bet
bet_amount = 1  # Betting $1 each time

# Simulation
results = []
cumulative_balance = 0

for trial in range(num_trials):
    # Spin the roulette wheel (0-36)
    spin_result = random.randint(0, 36)
    
    # Determine win or loss
    if spin_result == bet_number:
        cumulative_balance += bet_amount * payout_multiplier
    else:
        cumulative_balance -= bet_amount
    
    # Record the cumulative balance
    results.append(cumulative_balance)

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(range(1, num_trials + 1), results, marker='o', linestyle='-')
plt.title("Roulette Simulation: Betting on Number 36")
plt.xlabel("Trial Number")
plt.ylabel("Cumulative Balance ($)")
plt.grid()
plt.show()
```

```python
# Keno
import random
import matplotlib.pyplot as plt

# Parameters
num_games = 100
player_numbers = [54, 43, 26, 50, 5, 59, 71, 15, 29, 68]  # Your chosen numbers
payout_structure = {0: 0, 1: 0, 2: 0, 3: 0, 4: 1, 5: 8, 6: 40, 7: 100, 8: 300, 9: 1000, 10: 5000}  # Example payouts
payouts = []

# Simulate 100 games
for _ in range(num_games):
    # Casino draws 10 random numbers
    drawn_numbers = random.sample(range(1, 81), 10)
    # Count matches
    matches = len(set(player_numbers) & set(drawn_numbers))
    # Calculate payout
    payouts.append(payout_structure.get(matches, 0))

# Plot histogram of payouts
plt.figure(figsize=(10, 6))
plt.hist(payouts, bins=range(0, max(payouts) + 1), edgecolor='black', align='left', rwidth=0.8)
plt.title("Histogram of Payouts Over 100 Games")
plt.xlabel("Payout ($)")
plt.ylabel("Frequency")
plt.grid(axis='y')
plt.show()
```

```python
import random
import matplotlib.pyplot as plt

# Parameters
num_games = 100
player_numbers = [54, 43, 26, 50, 5, 59, 71, 15, 29, 68]  # Your chosen numbers
payout_structure = {0: 0, 1: 0, 2: 0, 3: 0, 4: 1, 5: 8, 6: 40, 7: 100, 8: 300, 9: 1000, 10: 5000}  # Example payouts
payouts = []

# Simulate 100 games
for _ in range(num_games):
    # Casino draws 10 random numbers
    drawn_numbers = random.sample(range(1, 81), 10)
    # Count matches
    matches = len(set(player_numbers) & set(drawn_numbers))
    # Calculate payout
    payouts.append(payout_structure.get(matches, 0))

# Calculate frequency of each payout
unique_payouts = sorted(set(payouts))
frequencies = [payouts.count(p) for p in unique_payouts]

# Plot histogram with separated bars
plt.figure(figsize=(10, 6))
plt.bar(unique_payouts, frequencies, edgecolor='black', width=1.0)
plt.title("Histogram of Payouts Over 100 Games")
plt.xlabel("Payout ($)")
plt.ylabel("Frequency")
plt.xticks(unique_payouts)  # Ensure all payout amounts are labeled
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
